## Outliers Removing

```python
def remove_outliers_iqr(data, columns):
    df_clean = data.copy()
    for col in columns:
        Q1 = df_clean[col].quantile(0.25)
        Q3 = df_clean[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        df_clean = df_clean[(df_clean[col] >= lower) & (df_clean[col] <= upper)]
    return df_clean

df_clean = remove_outliers_iqr(df, ['three_g','fc','px_height','m_dep'])
df=df_clean.copy()
```

# KNN Model

Mengklasifikasikan data berdasarkan voting mayoritas dari tetangga terdekat menggunakan perhitungan jarak. Implementasi metode ini dilakukan dengan langkah:

1. Menghitung **Euclidean Distance** antara data uji ($x$) dengan seluruh data latih ($x_i$): $$d(x, x_i) = \sqrt{\sum (x_j - x_{ij})^2}$$
2. Mengurutkan jarak dari terkecil ke terbesar.
3. Mengambil *K* tetangga terbesar.
4. Melakukan **Perhitungan Mayoritas** untuk menentukan kelas prediksi

## Class KNN

```python
class K_Nearest_Neighbors_Classifier:
    
    def __init__(self, K):
        self.K = K

    def fit(self, X_train, y_train):
        self.X_train = X_train.to_numpy()
        self.y_train = y_train.to_numpy()

        self.m, self.n = self.X_train.shape
        self.classes_ = np.unique(self.y_train)

    def euclidean(self, x1, x2):
        return np.sqrt(np.sum((x1 - x2) ** 2))

    def find_neighbors(self, x):
        distances = np.zeros(self.m)

        for i in range(self.m):
            distances[i] = self.euclidean(x, self.X_train[i])

        idx = np.argsort(distances)
        return self.y_train[idx][:self.K]

    def predict(self, X_test):
        X_test = X_test.to_numpy()
        y_pred = np.zeros(X_test.shape[0])

        for i, x in enumerate(X_test):
            neighbors = self.find_neighbors(x)
            y_pred[i] = mode(neighbors, keepdims=True)[0][0]

        return y_pred

    def predict_proba(self, X_test):
        X_test = X_test.to_numpy()
        proba = []

        for x in X_test:
            neighbors = self.find_neighbors(x)
            probs = np.zeros(len(self.classes_))

            for idx, cls in enumerate(self.classes_):
                probs[idx] = np.sum(neighbors == cls) / self.K

            proba.append(probs)

        return np.array(proba)
```

## Model Training and Prediction

```python
model_KNN = K_Nearest_Neighbors_Classifier(K=5)
model_KNN.fit(X_train, y_train)

y_pred = model_KNN.predict(X_val)
y_proba = model_KNN.predict_proba(X_val)
```

## Comparison with sklearn

```python
model_nn=KNeighborsClassifier(n_neighbors=5)
model_nn.fit(X_train, y_train)
y_pred=model_nn.predict(X_val)
```

```python
def evaluate_model(model, X, y, dataset_name="Dataset"):
    y_pred = model.predict(X)
    y_proba = model.predict_proba(X)
    
    accuracy = accuracy_score(y, y_pred)
    precision = precision_score(y, y_pred, average='weighted')
    recall = recall_score(y, y_pred, average='weighted')
    f1 = f1_score(y, y_pred, average='weighted')
    auc = roc_auc_score(y, y_proba, multi_class='ovr')
    
    print(f"\n===== {dataset_name} Evaluation =====")
    print(f"Accuracy: {accuracy * 100:.2f}%")
    print(f"Precision: {precision * 100:.2f}%")
    print(f"Recall: {recall * 100:.2f}%")
    print(f"F1 Score: {f1 * 100:.2f}%")
    print(f"ROC AUC: {auc * 100:.2f}%")
```

Untuk mengukur keberhasilan model pada dataset *multiclass* yang tidak seimbang (*imbalanced*), digunakan metrik sebagai berikut:
1.  **Akurasi (*Accuracy*):** Rasio prediksi benar terhadap keseluruhan data.
2.  **Precision, Recall, & F1-Score:** Digunakan untuk melihat performa detail per kelas obat, memastikan model tidak bias terhadap kelas mayoritas (DrugY).
3.  **ROC AUC:** Mengukur kemampuan model membedakan tiap kelas dengan pendekatan one vs rest. Nilai mendekati 1 menunjukkan pemisahan kelas yang baik pada data multiclass tidak seimbang.

# Hyperparameter Tuning with Genetic Algorithm

Alih-alih menggunakan metode *Grid Search* (mencoba satu per satu), parameter model dioptimalkan menggunakan Algoritma Genetika dengan komponen berikut:
1.  **Individu (Kromosom):** Merepresentasikan nilai parameter (Angka $K$ atau *Depth*).
2.  **Populasi:** Sekumpulan individu acak yang dibangkitkan pada awal proses.
3.  **Fitness Function:** Menggunakan nilai **Akurasi Validasi**. Semakin tinggi akurasi model dengan parameter tersebut, semakin tinggi nilai *fitness*-nya.
4.  **Seleksi & Crossover:** Dua individu terbaik dipilih menjadi induk, lalu dikawinkan (rata-rata nilai parameter) untuk menghasilkan keturunan.
5.  **Mutasi:** Modifikasi acak kecil ditambahkan pada parameter anak untuk menjaga variasi genetik.

```python
POP_SIZE = 20
GENERATIONS = 30
MUTATION_RATE = 0.2
ELITE_SIZE = 1   #ELITISM

N_NEIGHBORS_RANGE = (1, 20)
WEIGHTS_OPTIONS = ['uniform', 'distance']

def create_chromosome():
    return {
        'n_neighbors': random.randint(*N_NEIGHBORS_RANGE),
        'weights': random.choice(WEIGHTS_OPTIONS)
    }

def fitness(chromosome):
    model = KNeighborsClassifier(**chromosome)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    return accuracy_score(y_test, y_pred)

def tournament_selection(population, fitnesses, k=3):
    candidates = random.sample(list(zip(population, fitnesses)), k)
    return max(candidates, key=lambda x: x[1])[0]

def crossover(parent1, parent2):
    return {
        'n_neighbors': parent1['n_neighbors'] if random.random() < 0.5 else parent2['n_neighbors'],
        'weights': parent1['weights'] if random.random() < 0.5 else parent2['weights']
    }

def mutate(chromosome):
    if random.random() < 0.5:
        chromosome['n_neighbors'] = random.randint(*N_NEIGHBORS_RANGE)
    else:
        chromosome['weights'] = random.choice(WEIGHTS_OPTIONS)
    return chromosome

population = [create_chromosome() for _ in range(POP_SIZE)]
best_accuracies = []
for gen in range(GENERATIONS):
    fitnesses = [fitness(ind) for ind in population]

    elite_indices = np.argsort(fitnesses)[-ELITE_SIZE:]
    elites = [population[i] for i in elite_indices]

    new_population = elites.copy()

    while len(new_population) < POP_SIZE:
        parent1 = tournament_selection(population, fitnesses)
        parent2 = tournament_selection(population, fitnesses)

        child = crossover(parent1, parent2)

        if random.random() < MUTATION_RATE:
            child = mutate(child)

        new_population.append(child)

    population = new_population

    best_fitness = max(fitnesses)
    best_individual = population[np.argmax(fitnesses)]
    best_accuracies.append(best_fitness)
    print(f"Gen {gen+1}: Best Accuracy = {best_fitness:.4f}, Params = {best_individual}")

fitnesses = [fitness(ind) for ind in population]
best_idx = np.argmax(fitnesses)
best_solution = population[best_idx]

print("\nBest Hyperparameters Found:")
print(best_solution)
print(f"Accuracy:, {round(fitness(best_solution),2)}")
```

## Train Model with Genetic Algorithm Parameter

```python
KNN_GA=KNeighborsClassifier(
    n_neighbors=best_solution['n_neighbors'],
    weights=best_solution['weights']
)
KNN_GA.fit(X_train,y_train)
y_pred=KNN_GA.predict(X_test)
```

## Evaluation

## Evaluation Test Data

```python
evaluate_model(KNN_GA, X_test, y_test, "KNN with genetic algorithm parameters")
print(classification_report(y_test, y_pred, digits=4))
```

## Genetic Algorithm Training Convergence 

```python
plt.figure(figsize=(8,5))
plt.plot(range(1, GENERATIONS+1), best_accuracies, marker='o')
plt.title("Convergence of GA (Best Accuracy per Generation)")
plt.xlabel("Generation")
plt.ylabel("Best Accuracy")
plt.grid(True)
plt.show()
```
